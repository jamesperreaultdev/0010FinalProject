```python
#imports
import pandas
import matplotlib.pyplot as plt

#name the dataset of taxes chips for no reason
chip = pandas.read_csv("taxes.csv")
```

```python
#dictonary to store tax avoidance per hood
items = {}

#convert float values to ints to work better with the graphing library
check = chip['current_delq'].astype(int) 

#iterate through all rows in the dataframe
for row in chip.index: 
    #if the neighborhood in the row already is in the dictionary, add the value of it's tax deliquency 
    #to it's total.
    if (chip["neighborhood"].values[row]) in items:
        items[chip["neighborhood"].values[row]] += int((check.values[row]))
    #otherwise, make a new key and add this row's tax deliquency    
    else:
        items[(chip["neighborhood"].values[row])] = int((check.values[row]))
        
#print out the dictonary to check for testing
print(items)   

#find max value of tax deliquency in the dictionary, then get it's key (it's hood)
maxtax = max(items, key=items.get)

#print out the best neighhorhood
print("The neighborhood with the best tax deliquency is "+ maxtax)
```

```python
#plot a graph with a length of items, and rotate x bar names vertically for readability
plt.bar(range(len(items)), items.values(), align='center')
plt.xticks(range(len(items)), list(items.keys()), rotation='vertical')


plt.show()
```
