```python
import pandas as pd
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter


parksFile= pd.read_csv("parksPitt.csv")

neighborHoodPark = dict()

neighborHoods = []

numberOfParks = []


# Calculate frequency of parks in each neigborhood.
for index, row in parksFile.iterrows(): 
    park = row['type']
    
    if park == "Park": 
        
        neighborhood = row['neighborhood']
        
        if (neighborhood not in neighborHoodPark):
            neighborHoodPark[neighborhood] = 1 
        else: 
            neighborHoodPark[neighborhood] += 1 
    
# Make a bar chart of the amount of parks in each neighborhood, maybe with some nice pictures

# pd.Series([1,5,2,5], index=["cats", "dogs", "chickens", "spiders"]).plot.bar()

for neighborhood in neighborHoodPark.keys(): 
    
    neighborHoods.append(neighborhood)
    
for number in neighborHoodPark.values(): 
    
    numberOfParks.append(number)
    
x = pd.Series(numberOfParks, index = neighborHoods)
x.ylabel = "Neighborhoods"
x.xlabel = "Parks"
x.plot(kind = 'barh', figsize = (14, 14))





# Write a conclusion about your submetric, commit repo. 


```
